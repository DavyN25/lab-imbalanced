## 1. What is the distribution of our target variable? Can we say we're dealing with an imbalanced dataset?

```python
fraud.isnull().sum()
```

#### Conclusion:

## 2. Train a LogisticRegression.

#### Split features/target

```python
X = fraud.drop(columns=['fraud']) 
y = fraud['fraud']
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled_np = scaler.transform(X_train)
X_test_scaled_np = scaler.transform(X_test)
```

```python
log_reg = LogisticRegression()
```

```python
X_train_scaled_df = pd.DataFrame(X_train_scaled_np, columns=X_train.columns, index=X_train.index)
X_test_scaled_df  = pd.DataFrame(X_test_scaled_np, columns=X_test.columns, index=X_test.index)
```

```python
log_reg.fit(X_train_scaled_df, y_train)
```

## 3. Evaluate your model. Take in consideration class importance, and evaluate it by selection the correct metric.

```python
y_pred_test_log = log_reg.predict(X_test_scaled_df)
print(classification_report(y_pred_test_log, y_test))
```

```python
# Confusion Matrix :
print(confusion_matrix(y_test, y_pred_test_log))
```

Interpretation
The model is very effective at catching fraud (high recall = 0.89). 
But it generates a lot of false alarms (precision = 0.60), which could lead to unnecessary transaction blocks or manual reviews.
The F1-score of 0.72 for fraud shows a good balance between precision and recall.

Confusion Matrix : 
False negatives (missed frauds): 8,637
False positives (false alarms): 1,610

## 4. Run Oversample in order to balance our target variable and repeat the steps above, now with balanced data. Does it improve the performance of our model?


```python
train = pd.DataFrame(X_train_scaled_np, columns=X_train.columns, index=X_train.index)
```

```python
train["fraud"] = y_train.values
```

```python
# create 2 DF fraud and no fraud
fraud = train[train["fraud"] == 1]
no_fraud = train[train["fraud"] == 0]
```

```python
len(fraud),len(no_fraud)
```

```python
# resample to increase the number of the fraud class to have the same amount than the no fraud 65676 => 684324
yes_oversampled = resample(fraud,
                                    replace=True,
                                    n_samples = len(no_fraud),
                                    random_state=0)
```

```python
# combine the both df 
train_over = pd.concat([yes_oversampled, no_fraud])
train_over
```

```python
fraud_plt = train_over["fraud"].value_counts()
fraud_plt.plot(kind="bar")
plt.show()
```

```python
len(fraud),len(no_fraud)
```

```python
# repeat the steps above, now with balanced data.
X_train_over = train_over.drop(columns = ["fraud"])
y_train_over = train_over["fraud"]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train_over, y_train_over)
```

```python
#Make the prediction  and compare with the result we obtain before the oversampling
y_pred_test_log = log_reg.predict(X_test_scaled_df)
print(classification_report(y_pred = y_pred_test_log, y_true = y_test))
```

```python
# Confusion Matrix :
print(confusion_matrix(y_test, y_pred_test_log))
```

Interpretation :

Before oversampling: The model misses ~8,600 frauds
After oversampling: Only ~1,100 frauds missed
→ That’s a 7,500 fraud reduction, which is huge.
Trade-off: more false alarms (from 1,600 → 15,000), which may increase manual reviews or customer friction.

Oversampling significantly improves fraud detection (recall).
Even though precision drops and false positives increase, this trade-off is acceptable — especially in high-risk domains like fraud prevention.

## 5. Now, run Undersample in order to balance our target variable and repeat the steps above (1-3), now with balanced data. Does it improve the performance of our model?

```python
train
```

```python
# Now we want the sample not survived to be the same as the one survived. we reduice the no survived DF
no_undersampled = resample(no_fraud,
                                    replace=False,
                                    n_samples = len(fraud),
                                    random_state=0)
no_undersampled
```

```python
train_under = pd.concat([no_undersampled, fraud])
train_under
```

#### Separating our features from our target

```python
X_train_under = train_under.drop(columns = ["fraud"])
y_train_under = train_under["fraud"]
```

#### Model training

```python
log_reg = LogisticRegression()
log_reg.fit(X_train_under, y_train_under)
```

#### Evaluate our model

```python
y_pred_test_log = log_reg.predict(X_test_scaled_df)
print(classification_report(y_pred = y_pred_test_log, y_true = y_test))
```
